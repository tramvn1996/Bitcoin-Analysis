```python
x = data['date1']
y = data['btc_price']
y_eth = data['eth_price']
y_ltc = data["ltc_price"]

fig = plt.figure()
fig = plt.figure(figsize=(20,12))

fig = plt.plot(x,y,label = "BTC", linewidth=0.5) #thêm legend parameter
fig = plt.plot(x,y_eth,label = "ETH", linewidth=0.5)
fig = plt.plot(x,y_ltc,label="LTC", linewidth=0.5)
fig =plt.xlabel("Date")
fig = plt.xticks(rotation='20')
fig = plt.ylabel("price(USD)")
fig =plt.title("Price of Bitcoin, Ethereum, and Litecoin")
fig = plt.legend() #thêm hàm legend()
fig = plt.savefig("BTL_na.png")
plt.show()
```

```python
data = pd.read_csv("../Data2/ver4.csv")
```

```python
data.head()
```

```python
data['date'] =pd.to_datetime(data['date'])
```

```python
x = data['date']
y = data['price(USD)']
y_eth = data['txCount']/10
y_ltc = data["btc_unweighted"]

fig = plt.figure()
fig = plt.figure(figsize=(20,12))

fig = plt.plot(x,y,label = "BTC", linewidth=0.7) #thêm legend parameter
fig = plt.plot(x,y_eth,label = "Transaction counts", linewidth=0.7)

fig =plt.xlabel("Date")
fig = plt.xticks(rotation='20')
fig = plt.ylabel("price(USD)")
fig =plt.title("Price of Bitcoin vs Transactions in BTC")
fig = plt.legend() #thêm hàm legend()
fig = plt.savefig("BTL_txcount.png")
plt.show()
```
